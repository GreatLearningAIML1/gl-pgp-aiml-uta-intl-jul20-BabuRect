# Distribution plot of predictor variables for each class of the target

```python
# Split the dataframe with Target = yes and Target = no
TargetYesDf = copyBankDf[copyBankDf['Target'] == 1]
TargetNoDf  = copyBankDf[copyBankDf['Target'] == 0]

# Get all predictor variables
colList = TargetYesDf.columns.tolist()
colList.remove('Target')

# Split the continous and categorical
ContList = colList[0:5]
catList  = colList[5:]

for col in ContList:
    sns.distplot(TargetYesDf[col], hist=False,label='yes')
    sns.distplot(TargetNoDf[col], hist=False,label='No')
    print("Dist plot of ",col)
    plot.legend()
    plot.show()   
    
# Columns with categorical is shown in distplot with a bandwidth.Otherwise it throws error. Thatswhy
# categorical columns are splitted
for col in catList:
    sns.distplot(TargetYesDf[col], hist=False,label='yes',kde_kws={'bw':0.1})
    sns.distplot(TargetNoDf[col], hist=False,label='No',kde_kws={'bw':0.1})
    print("Dist plot of ",col)
    plot.legend()
    plot.show()

```

# Deliverable - 3
## Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression as LogReg
logModel = LogReg( solver='liblinear')

# fit the model
logModel.fit(X_train,Y_train)

# get the coefficient of variables and print
coeff = pd.DataFrame(logModel.coef_)
print(coeff)

# get the model and score and print it
score = logModel.score(X_test,Y_test)
print("-------------------------------------------------")
print("Logistic regression score on test data = ", score)
print("-------------------------------------------------")
```

```python
# Function to get the scores given observed data, predicated data. Called by multiple models
from sklearn.metrics import confusion_matrix, recall_score, precision_score, f1_score, roc_auc_score,accuracy_score

def ModelMetrics(model,trainScore, testScore, observed, predicted, df):
    
    # Print the various score
    print("-------------------------------------------")
    print("Scores of model: ",model)
    print("-------------------------------------------")
    print("Train data Score =", trainScore)
    print("Test  data Score =", testScore)
    
    rcScore  = recall_score(observed, predicted)
    prScore  = precision_score(observed, predicted)
    rocScore = roc_auc_score(observed, predicted)
    acc      = accuracy_score(observed, predicted)
    f1Score  = f1_score(observed, predicted)
    
    print("Recall Score     =", rcScore)
    print("Precision Score  =", prScore)
    print("AUC Score        =", rocScore)
    print("Accuracy Score   =", acc)
    print("F1 Score         =", f1Score)
    
    df = df.append({'Model':model,
                       'Train Score':trainScore,
                       'Test Score': testScore,
                       'Recall':rcScore,
                       'Precision':prScore,
                       'Roc Score':rocScore,                       
                       'accuracy': acc,
                       'f1 score': f1Score},ignore_index=True)
    
    # Draw the confusion matrix
    confMat = confusion_matrix(observed,predicted)
    sns.heatmap(confMat, annot=True,  fmt='.2f', xticklabels = [0,1] , yticklabels = [0,1] )
    plot.ylabel('Observed')
    plot.xlabel('Predicted')
    plot.show()
    
    return df
```

```python
y_predict = logModel.predict(X_test)    
```

```python
# Score of the decision tree
trainScore = logModel.score(X_train,Y_train)
testScore  = logModel.score(X_test,Y_test)
```

```python
# Get the various metrics of the model
metricDf = ModelMetrics("Logistic Regression",trainScore,testScore, Y_test,y_predict,metricDf)
```

## Bagging

```python
from sklearn.ensemble import BaggingClassifier
bagClas = BaggingClassifier(n_estimators=40, random_state=1)
bagClas.fit(X_train,Y_train)

trainScore = bagClas.score(X_train,Y_train)
testScore  = bagClas.score(X_test,Y_test)

y_predict_bagCl = bagClas.predict(X_test)
metricDf = ModelMetrics("Bagging Classifier",trainScore,testScore,Y_test,y_predict_bagCl,metricDf)
```

## Bagging uses complex model. And the accuracy is improved little bit than the decision tree but almost as same as the regularized decision tree.
